```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

#dados que serão utilizados
url = "https://mantis.saojudas.base2.com.br/login_page.php"

plataforma = "teste"
plataforma_editada = "teste editar"
so = "operacional"
versao = "1"
descricao = "descrição do so"
```

```python
#realizar login 
def setup(navegador):
    #realiza login
    navegador.find_element_by_xpath('//*[@id="username"]').send_keys('Fernanda')
    navegador.find_element_by_xpath('//*[@id="login-form"]/fieldset/input[2]').click()
    navegador.find_element_by_xpath('//*[@id="password"]').send_keys(81721367)
    navegador.find_element_by_xpath('//*[@id="login-form"]/fieldset/input[3]').click()
    assert "https://mantis.saojudas.base2.com.br/my_view_page.php" in navegador.current_url
```

```python
def criar_perfil_so(): 
    #abre o navegador e acessa a url
    navegador = webdriver.Chrome()
    navegador.get(url)

    #realiza o login
    setup(navegador)

    #clicar no nome de usuario
    navegador.find_element_by_xpath('//*[@id="navbar-container"]/div[2]/ul/li[2]/a').click()

    #clicar em Minha Conta
    navegador.find_element_by_xpath('//*[@id="navbar-container"]/div[2]/ul/li[2]/ul/li[1]/a').click()

    #verifica se acessou a pagina
    assert 'https://mantis.saojudas.base2.com.br/account_page.php' in navegador.current_url

    #clicar em Perfis
    navegador.find_element_by_xpath('//*[@id="main-container"]/div[2]/div[2]/div/ul/li[4]/a').click()

    #verifica se acessou a pagina
    assert 'https://mantis.saojudas.base2.com.br/account_prof_menu_page.php' in navegador.current_url

    #preencher campos e criar perfil
    navegador.find_element_by_id('platform').send_keys(plataforma)
    navegador.find_element_by_id('os').send_keys(so)
    navegador.find_element_by_id('os-version').send_keys(versao)
    navegador.find_element_by_id('description').send_keys(descricao)

    navegador.find_element_by_xpath('//*[@id="account-profile-form"]/fieldset/div/div[2]/div[2]/input').click()
    
    #acessa pag Criar Tarefas
    navegador.find_element_by_xpath('//*[@id="sidebar"]/ul/li[3]/a').click()
    
    #verifica select de Perfil para ver se SO criado está disponível
    element = plataforma + " " + so + " " + versao
    lista = []
    try:
        if navegador.find_element_by_css_selector('#profile_closed_link').is_displayed():
            navegador.find_element_by_css_selector('#profile_closed_link').click()
        WebDriverWait(navegador, 10).until(
            EC.visibility_of(navegador.find_element_by_xpath('//*[@id="profile_open"]/table'))
        )
    finally:
        #carrega opcoes do select em uma lista
        select = Select(navegador.find_element_by_id('profile_id'))
        options = select.options
        for e in options:
            lista.append(e.text)
            
    #verifica se o perfil criado está na lista
    assert element in lista
    
    navegador.close()
    print("Criar Perfil de SO - OK")

criar_perfil_so()
```
